#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO
import matplotlib.pyplot as plt


# Función para normalizar los datos
def normalizar(x):
    return (x - np.mean(x)) / np.std(x)

# Crear un nuevo DataFrame para almacenar los datos normalizados
df_norm = ds_filtrado.copy()

# Iterar sobre los grupos y normalizar la variable X
for p in poblaciones:
    df_norm.loc[df_norm['level5'] == p, 'price_norm'] = normalizar(ds_filtrado['price'])



# Crear un histograma para cada grupo
for p in poblaciones:
    plt.hist(df_norm[df_norm['level5'] == p]['price_norm'], 
             bins=10, alpha=0.5, label=p)

# Personalizar el gráfico
plt.xlabel('precio normalizado')
plt.ylabel('Frecuencia')
plt.title('Histogramas de precios por grupo')
plt.legend()
plt.show()

```

**Analisis**. 


Se utilizo la normalización en base a la media y desvición estandar de los datos intergrupo que definen el cinturon sur de Madrid, la cual consiste en tomar cada dato dentro del grupo de interes y restarle su media y desviación estandar.
